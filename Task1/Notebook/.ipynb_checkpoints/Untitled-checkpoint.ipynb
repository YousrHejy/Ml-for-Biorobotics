```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt 
  
# Linear Regression 
class LinearRegression() :
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate  
        self.iterations = iterations 
        self.weights = None
        self.bias = None
          
    # Function for model training 
    def fit( self, X, Y ) :
        # no_of_training_examples, no_of_features   
        self.no_samples, self.no_features = X.shape 
        # weight initialization 
        self.weights = np.zeros(self.no_features) 
        self.bias = 0
        self.X = X 
        self.Y = Y 
        
        # predicting then gradient descent learning          
        for i in range( self.iterations ):   
            self.gd_update_weights() 
        return self
      
    # Function to update weights in gradient descent 
    def gd_update_weights(self) :  
#         Y_pred = self.predict(self.X) 
        # calculate gradients   
#         dW = - ( 2 * ( self.X.T ).dot( self.Y - Y_pred )  ) / self.m 
        dw = (1/self.no_samples) * (np.dot(self.X.T, (np.dot(self.weights, self.X) - self.Y)))
#         db = - 2 * np.sum( self.Y - Y_pred ) / self.rows  
        db = (1/self.no_samples) * (np.sum(np.dot(self.X,self.weights) - self.Y))
        # update weights 
        self.weights = self.weights - self.learning_rate * dw
        self.bias = self.bias - self.learning_rate * db 
        return self
      
    # function of linear regression (Y = XW)
    def predict(self, X): 
        y_pred = np.dot(X, self.weights) + self.bias
        return y_pred 
```

```python
 # Importing dataset 
df = pd.read_csv( "insurance.csv" ) 
df
# X = df.drop(['charges'], axis=1)
# y = df['charges']
# X = df.iloc[:, :6]
X = df.values[:, :-1]
y = df.values[:, -1]
type(X)
y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)   

# Model training 
model = LinearRegression(iterations = 1000, learning_rate = 0.01) 
model.fit(X_train, y_train ) 
```
